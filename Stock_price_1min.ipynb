```python
from tvDatafeed import TvDatafeed, Interval
```

```python
tv = TvDatafeed()
```

```python
Ticker='INDUSINDBK'
Market='NSE'
```

```python

df= tv.get_hist(symbol=Ticker,
                exchange=Market,
                interval=Interval.in_1_minute,
                n_bars= 5000,
                extended_session=False
                )
```

```python
df
```

```python
df=df.reset_index()
df=df[['datetime','open','high','low','close','volume']]
```

```python
df.columns=['Time','Open','High','Low','Close','Volume']
```

```python
df = df.sort_values(by=['Time'])
```

```python
df['Time'] = df['Time'].dt.tz_localize('GMT').dt.tz_convert('Asia/Kolkata').dt.strftime('%m/%d/%Y %H:%M')
```

```python
df
```

```python
min_date = df['Time'].min()
max_date = df['Time'].max()
print(min_date,max_date)
```

```python
filename = f"{Ticker}.csv"
df.to_csv(f"D:\Projects\{filename}", index=False)
```
